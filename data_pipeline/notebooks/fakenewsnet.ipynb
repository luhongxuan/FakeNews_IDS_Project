```python
import pandas as pd

# FakeNewsNet 有兩個來源：politifact 和 gossipcop
politifact_fake = pd.read_csv("data/raw/FakeNewsNet/dataset/gossipcop_fake.csv")
politifact_real = pd.read_csv("data/raw/FakeNewsNet/dataset/gossipcop_real.csv")

politifact_fake["label"] = "fake"
politifact_real["label"] = "real"

df = pd.concat([politifact_fake, politifact_real], ignore_index=True)

print(f"總新聞數：{len(df)}")
print(f"欄位：{df.columns.tolist()}")
df.head()
```

```python
print("=== fake vs real ===")
print(df["label"].value_counts())

print("\n=== 缺失值 ===")
print(df.isnull().sum())

print("\n=== 標題長度分佈 ===")
df["title_len"] = df["title"].str.len()
print(df["title_len"].describe())
```
